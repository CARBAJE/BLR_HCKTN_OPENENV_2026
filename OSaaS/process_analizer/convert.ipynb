```python
import json

def mapear_instrucciones(archivo_original, archivo_con_titulos, archivo_salida):
    try:
        # 1. Cargar los datos de ambos archivos
        with open(archivo_original, 'r', encoding='utf-8') as f:
            datos_originales = json.load(f)
            
        with open(archivo_con_titulos, 'r', encoding='utf-8') as f:
            datos_con_titulos = json.load(f)

        # 2. Iterar y mapear las instrucciones
        # Usamos zip para recorrer ambos al mismo tiempo
        for i, (paso_orig, paso_titulo) in enumerate(zip(datos_originales, datos_con_titulos)):
            if "instruction" in paso_titulo:
                paso_orig["instruction"] = paso_titulo["instruction"]
            else:
                print(f"Advertencia: El paso {i} en el archivo de títulos no tiene instrucción.")

        # 3. Guardar el resultado en un nuevo archivo
        with open(archivo_salida, 'w', encoding='utf-8') as f:
            json.dump(datos_originales, f, indent=2, ensure_ascii=False)
            
        print(f"Éxito: Se han mapeado las instrucciones en '{archivo_salida}'")

    except FileNotFoundError as e:
        print(f"Error: No se encontró el archivo - {e}")
    except json.JSONDecodeError:
        print("Error: Uno de los archivos no tiene un formato JSON válido.")
    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")

# Configuración de nombres de archivo
mapear_instrucciones(
    'real.json',      # El que tiene "(sin instrucción)"
    'names.json',       # El que tiene las frases reales
    'resultado_final.json' # El archivo que se creará
)
```
